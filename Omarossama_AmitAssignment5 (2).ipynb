```python
 class Bank:
    users = [
        {'name': 'admin', 'password': "123", 'balance': 1234},
       ]

    def __init__(self):
        print("Welcome")
        self.menu1()

    def user_register(self):
        username = input("Please enter username: ")
        while any(user['name'] == username for user in self.users):
            print("Username already exists")
            username = input("Please enter username: ")
        password = input("Enter a password: ")
        self.users.append({"name": username, "password": password, "balance": 0})

    def user_login(self):
        username = input("Please enter username: ")
        password = input("Please enter password: ")
        while not any(user['name'] == username and user["password"]==password for user in self.users):
            username = input("Please enter correct username: ")
            password=input("Enter password: ")
        user = self.getuser(username)
        print("Login Successfully\nWelcome Back ")
        self.menu2(user)
        
    def getuser(self,username):
        for user in self.users:
            if username in user["name"]:
                return user
             
    def user_deposit(self, user):
        deposit = int(input("Please enter the number you want to deposit: "))
        user['balance'] += deposit
        print(f"Your balance is now {user['balance']}")

    def user_withdraw(self, user):
        withdraw = int(input("Please enter the number you want to withdraw: "))
        if withdraw > user['balance']:
            print("Insufficient balance")
        else:
            user['balance'] -= withdraw             
            print(f"Your balance is now {user['balance']}")

    def user_check(self, user):
        print("Current balance:", user['balance'])  
    
    def menu1(self):
        while True:
            print("1. Register")
            print("2. Login")
            print("0. Close")
            x = input("Please choose an option: ")
            if x == '1':
                self.user_register()
            elif x == '2':
                self.user_login()
            elif x == '0':
                print("Goodbye")
                break
            else:
                print("Choose a valid option")

    def menu2(self, user):
        while True:
            print("1. Check balance")
            print("2. Deposit")
            print("3. Withdrawal")
            print("0. Return to previous menu")
            x = input("Please choose a number from the menu: ")
            if x == '1':
                self.user_check(user)
            elif x == '2':
                self.user_deposit(user)
            elif x == '3':
                self.user_withdraw(user)
            elif x == '0':
                break
            else:
                print("Choose valid option")
omar=Bank()                
```
